```python
# [셀 0] 기본 설정
from pathlib import Path
import shutil
import random
import os

BASE_DIR = Path("/content/drive/MyDrive/먼작귀")

# 새로 만든 YOLO 학습 데이터셋
DRIVE_DATASET = BASE_DIR / "dataset" / "yolo_scenario_dataset"
LOCAL_DATASET = Path("/content/yolo_scenario_dataset")

# 기존 개별상품 학습 모델 best.pt 경로
OLD_MODEL_PATH = BASE_DIR / "runs" / "product_detection" / "yolov8n_product-3" / "weights" / "best.pt"

# baseline 결과 저장 위치
BASELINE_PROJECT_DIR = BASE_DIR / "runs" / "baseline"

print("DRIVE_DATASET:", DRIVE_DATASET, DRIVE_DATASET.exists())
print("OLD_MODEL_PATH:", OLD_MODEL_PATH, OLD_MODEL_PATH.exists())
print("BASELINE_PROJECT_DIR:", BASELINE_PROJECT_DIR)
```

```python
# [셀 1] 기존 모델 baseline 테스트: front/s0
from ultralytics import YOLO

test_source_front = BASE_DIR / "dataset" / "synthetic" / "images" / "front" / "s0"

print("기존 모델 존재:", OLD_MODEL_PATH.exists(), OLD_MODEL_PATH)
print("front 테스트 폴더 존재:", test_source_front.exists(), test_source_front)
print("front 테스트 이미지 개수:", len(list(test_source_front.glob("*.png"))))

model = YOLO(str(OLD_MODEL_PATH))

baseline_front_results = model.predict(
    source=str(test_source_front),
    imgsz=1280,
    conf=0.25,
    save=True,
    save_txt=True,
    save_conf=True,
    project=str(BASELINE_PROJECT_DIR),
    name="baseline_eval_front_s0",
    exist_ok=True,
)
```

```python
# [셀 3] 기존 모델 baseline 테스트: side/s0
test_source_side = BASE_DIR / "dataset" / "synthetic" / "images" / "side" / "s0"

print("side 테스트 폴더 존재:", test_source_side.exists(), test_source_side)
print("side 테스트 이미지 개수:", len(list(test_source_side.glob("*.png"))))

baseline_side_results = model.predict(
    source=str(test_source_side),
    imgsz=1280,
    conf=0.25,
    save=True,
    save_txt=True,
    save_conf=True,
    project=str(BASELINE_PROJECT_DIR),
    name="baseline_eval_side_s0",
    exist_ok=True,
)
```

```python
# [셀 4] baseline 결과 요약
def summarize_yolo_predict_results(results, name):
    image_count = len(results)
    pred_count_list = []

    for r in results:
        if r.boxes is None:
            pred_count_list.append(0)
        else:
            pred_count_list.append(len(r.boxes))

    total_pred = sum(pred_count_list)
    avg_pred = total_pred / max(1, image_count)

    print(f"\n[{name}]")
    print("이미지 수:", image_count)
    print("총 탐지 bbox 수:", total_pred)
    print("이미지당 평균 bbox 수:", round(avg_pred, 2))
    print("최소 bbox 수:", min(pred_count_list) if pred_count_list else 0)
    print("최대 bbox 수:", max(pred_count_list) if pred_count_list else 0)

summarize_yolo_predict_results(baseline_front_results, "기존 모델 front/s0")
summarize_yolo_predict_results(baseline_side_results, "기존 모델 side/s0")
```
